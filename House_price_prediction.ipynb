```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
#  Load the dataset

url = "https://raw.githubusercontent.com/deep089/House_price_prediction/main/final_dataset.csv"
data = pd.read_csv(url)
```

```python
# Data Cleaning
data = data.dropna()
data = data[['size', 'price']]  # Keep only relevant columns
# Visualize the data
plt.figure(figsize=(10, 6))
sns.scatterplot(x='size', y='price', data=data)
plt.title('Square Footage vs House Price')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.grid(True)
plt.show()
```

```python

# Step 4: Simple Linear Regression
X = data['size'].values.reshape(-1, 1)
y = data['price'].values
model = LinearRegression()
model.fit(X, y)
```

```python
predictions = model.predict(X)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='size', y='price', data=data, label='Data Points')
plt.plot(data['size'], predictions, color='red', label='Regression Line')
plt.title('Linear Regression: Square Footage vs Price')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Intercept (a): {model.intercept_}")
print(f"Slope (b): {model.coef_[0]}")
```

```python
square_footage = 2000
predicted_price = model.predict(np.array([[square_footage]]))
print(f"Predicted price for a {square_footage} sq ft house: ${predicted_price[0]:,.2f}")
```
